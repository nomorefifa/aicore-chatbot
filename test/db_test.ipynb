```python
client = chromadb.PersistentClient(path='../data/vector_db')
col = client.get_collection('instructor_resumes')
print(f"총 청크 수: {col.count()}")

from collections import Counter

all_data = col.get(include=['metadatas'])
print("=== 강사별 ===")
for name, cnt in Counter(m['instructor_name'] for m in all_data['metadatas']).most_common():
    print(f"  {name}: {cnt}개")

print("\n=== 섹션별 ===")
for sec, cnt in Counter(m['section'] for m in all_data['metadatas']).most_common():
    print(f"  {sec}: {cnt}개")

result = col.get(
      where={'instructor_name': '김 승 현'},
      include=['documents', 'metadatas']
  )
for i in range(len(result['ids'])):
    print(f"섹션: {result['metadatas'][i]['section']}")
    print(f"내용: {result['documents'][i]}")  # 전체 출력
    print()

# 첫 번째 청크의 메타데이터 전체 확인
result = col.get(limit=1, include=['metadatas'])
print(result['metadatas'][0])
```
